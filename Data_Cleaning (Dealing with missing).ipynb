```python
import pandas as pd
import numpy as np
```

```python
dicti={'Temperature':[1,'NaN',3,2,3,1,2,'N/a',1,'na'],
      'Humidity':[1.0,'NaN',31.0,22.0,33.0,11.0,21.0,24.0,12.0,32.0]}
df=pd.DataFrame(dicti)
df.to_csv("datacleaning.csv")
```

# Step 1 : Detecting N/a,na,NaN values::

```python
df.isnull()
```

```python
missing_arr=['N/a','na',np.nan]
df1=pd.read_csv("datacleaning.csv",na_values=missing_arr)
```

```python
df1.isnull()
```

```python
#how to check if their are na values in df
df1.isnull().any()
```

# Step 2:  Remove these values(fill in places of those) :

```python
df1
```

```python
#drop all na avlues:
df1.dropna()
```

```python
#drop only those rows in which all column values=na
df1.dropna(how='all')  
```

```python
df1.dropna(how='any')
```

```python
#fill  in these values:
df1.fillna(0)
```

```python
#forward fill -- previous values get copied:
df1.fillna(method='ffill')
```

```python
#backward fill---next values get copies
df1.fillna(method='bfill')
```

```python
#interpolate : it takes the mean of above and below value /3 and fill it.
df1.interpolate()
```

```python
#specific colum filling
df1.fillna({'Temperature':999})
```
