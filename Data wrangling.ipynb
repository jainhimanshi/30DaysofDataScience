```python
#working with  csv data(excel-consist of rows and. columns)
#read a. csv data 
import pandas as pd
df=pd.read_csv("./Downloads/avocado.csv", encoding="utf-8")
df.head(2)
df["newcolumn"]=df["AveragePrice"].rolling(5).mean()
df.dropna().head(2)
df.to_csv("new_avocado")
 # we rename the column by replacing spaces with underscores and setting all characters to lowercase, so the 
#column names more closely resemble columns names one might find in a table.
new_data = pd.read_csv("./Downloads/avocado.csv")
new_data.rename(columns = lambda x: x.replace(' ', '_').lower(), inplace=True)
new_data.head(2)
```

```python
#working with. database:(its usually a collection of spreadsheet consisting of columns and rows )
#database schema : blueprint. that  tells us how we plan the database.
#SQL Syntax::
# (iyt will give all rows and columns of aadhar data)   SELECT * FROM data_aadhar  
#(if we havew to. set. a limit on number of rows we. need )  SELECT * FROM data_aadhar LIMIT 20;
#(selecting certain columns only)  SELECT columnname FROM data_aadhar LIMIT 20;
import pandasql 
q = """
        SELECT averageprice,total_volume
        FROM new_data
        LIMIT 2; 
        """ 

#Execute your SQL command against the pandas frame
#Execute your SQL command against the pandas frame
solution = pandasql.sqldf(q.lower(), locals())
solution
```

```python
#(selecting rows  with particular column values) SELECT * FROM data WHERE columnvalues="required value"
e="""
    SELECT * 
    FROM new_data
    WHERE  region="Albany"
    LIMIT 2;
    """
solution2=pandasql.sqldf(e,locals())
solution2
```

```python
#grouping by certain columnname.(SELECT * FROM data GROUPBY columnname)
r="""
   SELECT  date,region,sum(averageprice)
   FROM new_data
   GROUP BY date
   LIMIT 2;
"""
solution3=pandasql.sqldf(r,locals())
solution3
```
