# Loan Prediction-Example:


Problem Statement 

Dream Housing Finance company deals in all home loans. They have presence across all urban, semi urban and rural areas. Customer first apply for home loan after that company validates the customer eligibility for loan. Company wants to automate the loan eligibility process (real time) based on customer detail provided while filling online application form. These details are Gender, Marital Status, Education, Number of Dependents, Income, Loan Amount, Credit History and others. To automate this process, they have given a problem to identify the customers segments, those are eligible for loan amount so that they can specifically target these customers.
It is a classification problem where we have to predict whether a loan would be approved or not. In a classification problem, we have to predict discrete values based on a given set of independent variable(s). 

Loan prediction is a very common real-life problem that each retail bank faces atleast once in its lifetime. If done correctly, it can save a lot of man hours at the end of a retail bank.

We have to predict whether the loan will get approved or not?

```python
#Binning using pandas :
#When dealing with continuous numeric data, it is often helpful to bin the data into multiple buckets for further analysis. There are several different terms for binning including bucketing, discrete binning and discretization.

#Pandas supports these approaches using the cut and function. In this section you will learn how to use the pandas functions to convert continuous data to a set of discrete buckets.
```

```python
#Binning in Pandas with Age Example
#Create Random Age Data
#First, let's create a simple pandas DataFrame assigned to the variable df_ages with just one colum for age. This column will contain 8 random age values between 21 inclusive and 51 exclusive,

import numpy as np
df_ages = pd.DataFrame({'age': np.random.randint(21, 51, 8)})
df_ages
```

```python
#bucketing into age brackets : (20,29],(30,39],(40,49] ----see brackets
df_ages['age bucket']=pd.cut(x=df_ages['age'],bins=[20,29,39,49])
df_ages
```

```python
#handling missing values : fillna()--fill the missing places by mean/mode/meadian
#we have to import scipi for determining the mode/mean etc

from scipy.stats import mode
mode(df_ages['age'])

#This returns both mode and count. Remember that mode can be an array as there can be multiple values with high frequency. We will take the first one by default always using:
mode(df_ages['age']).mode[0]

#fill it 
df_ages['age'].fillna(mode(df_ages['age']).mode[0], inplace=True)
```

```python
#apply function :

#basically,whenever you want to iterate over df or its rows/columns.You will think of apply function . rathe rthan for loop as its faster also.
```

```python
#row wise  access
df_ages.apply(lambda x: x)
#first row access
df_ages.apply(lambda x:x[0])

#columnwise access:
df_ages.apply(lambda x:x,axis=1)
#age column access
df_ages.apply(lambda x: x['age'],axis=1)
```

# Merging Data using pandas

```python
#two functions which are used :
#concat()---when we have to append df
#merge()--when we have to merge on basis on common columns

df1=new_df['Outlet_Location_Type'].head()
df1
```

```python
df2=new_df['Outlet_Size'].head()
df2
```

```python
df3=new_df['Outlet_Type'].head()
df3
```

```python
#concat function appends the df to each other
result=pd.concat([df1,df2,df3])
result
```

```python
#pandas also offers us to provide keys in concat so that we know whioch df came from where:
result=pd.concat([df1,df2,df3],keys=['x','y','z'])
result
#we can retreive any df using reuslt.iloc("key") 
```

```python
#we can aslo join using column
#join='outer'===unio
result=pd.concat([df1,df2],axis=1,join='outer')
result
```

```python
#join='inner'====intersection
result=pd.concat([df1,df2],axis=1,join='inner')
result
```

```python
#merge() -----combines 2 df on basis on common column:
df1=new_df[['Outlet_Location_Type','Outlet_Size']].head()
df1
df2=new_df[['Outlet_Location_Type','Outlet_Size','Outlet_Type']].head()
df2
df3=new_df[['Outlet_Type','Outlet_Location_Type']].head()
df3
```

```python
#combine df1,df2 on basis on outlet location type
result=pd.merge(df1,df3,on='Outlet_Location_Type',how='outer')
result
```

```python
result=pd.merge(df1,df3,on='Outlet_Location_Type',how='inner')
result
```

```python
#alll from right+intersection of both
result=pd.merge(df1,df3,on='Outlet_Location_Type',how='right')
result
```

```python
#left join=left full +intersectrion
result=pd.merge(df1,df3,on='Outlet_Location_Type',how='right')
result
```
